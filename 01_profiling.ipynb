## 01_profilling
### Purpose: examine, analyse, and summarize raw data

```sql
USE curso.ventas;
```

### Examining fact and dimesion tables

```sql
DESCRIBE TABLE facturas_raw;
```

```sql
SELECT
  COUNT(*) AS filas,
  SUM(CASE WHEN num_factura IS NULL THEN 1 ELSE 0 END) AS null_reales,
  SUM(CASE WHEN num_factura IS NOT NULL AND TRIM(num_factura) = '' THEN 1 ELSE 0 END) AS vacios,
  SUM(CASE WHEN num_factura IS NOT NULL AND LOWER(TRIM(num_factura)) = 'null' THEN 1 ELSE 0 END) AS null_texto
FROM curso.ventas.facturas_raw;

```

```sql
SELECT
  COUNT(*) AS filas,
  SUM(CASE WHEN fecha_venta IS NULL THEN 1 ELSE 0 END) AS null_reales,
  SUM(CASE WHEN fecha_venta > current_date() THEN 1 ELSE 0 END) AS fecha_invalida,
  MIN(fecha_venta) AS min_fecha,
  MAX(fecha_venta) AS max_fecha
FROM curso.ventas.facturas_raw;
```

```sql
SELECT
  COUNT(*) AS filas,
  SUM(CASE WHEN producto IS NULL THEN 1 ELSE 0 END) AS null_reales
FROM curso.ventas.facturas_raw;
```

```sql
SELECT
  COUNT(*) AS filas,
  SUM(CASE WHEN cliente IS NULL THEN 1 ELSE 0 END) AS null_reales
FROM curso.ventas.facturas_raw;
```

```sql
SELECT
  COUNT(*) AS filas,
  SUM(CASE WHEN cantidad IS NULL THEN 1 ELSE 0 END) AS null_reales,
  SUM(CASE WHEN cantidad <= 0 THEN 1 ELSE 0 END) AS cantidad_invalida

FROM curso.ventas.facturas_raw;
```
